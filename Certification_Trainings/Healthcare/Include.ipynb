```
import requests

def get_home():
  user = dbutils.notebook.entry_point.getDbutils().notebook().getContext().tags().get("user").get()
  return f"/Users/{user}"

def downloadFile(url, dir):
  filename = url.split('/')[-1]
  path = dir+"/"+filename
  try:
    dbutils.fs.ls(path)
  except:
    dbutils.fs.mkdirs(dir)
    r = requests.get(url)
    with open("/dbfs"+path, 'wb') as f:
      f.write(r.content)
#  return (dbutils.fs.ls(path))
```
